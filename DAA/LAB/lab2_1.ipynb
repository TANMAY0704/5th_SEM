```python
import math
import matplotlib.pyplot as plt
```

```python
def method1(n_rs__69):
    count_rs_69 = 0
    for i in range(2, n_rs__69):
        count_rs_69 += 1
        if n_rs__69 % i == 0:
            return False, count_rs_69
    return True, count_rs_69
```

```python
def method2(n_rs_69):
    count_rs_69 = 0
    for i_rs_69 in range(2, n_rs_69 // 2):
        count_rs_69 += 1
        if n_rs_69 % i_rs_69 == 0:
            return False, count_rs_69
    return True, count_rs_69
    
    
    
```

```python
def method3(n_rs_69):
    count_rs_69 = 0
    for i in range(2, int(math.sqrt(n_rs_69)) + 1):
        count_rs_69 += 1
        if n_rs_69 % i == 0:
            return False, count_rs_69
    return True, count_rs_69
```

```python

inputs = [3, 7, 11, 15, 23, 50, 89, 97, 101, 109]
algorithms = [method1, method2 , method3 ]

loop_counts = {algo.__name__: [] for algo in algorithms}

for input_num in inputs:
    for algo in algorithms:
        _, count = algo(input_num)
        loop_counts[algo.__name__].append(count)

        print(f"Number: {input_num}, Algorithm: {algo.__name__}, Count: {count}")

# Plot the loop execution counts
plt.figure(figsize=(10, 6))
for algo in algorithms:
    plt.plot(inputs, loop_counts[algo.__name__], label=algo.__name__)
plt.xlabel('Input Numbers')
plt.ylabel('Number of Loop Iterations')
plt.title('Comparison of Loop Execution Counts for Prime Checking Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```
